```python
#三角形の面積を計算する
def calc_tri(base,height):
    return base * height * 0.5
```

try文の中でエラーが起きた際には、exceptに処理が変わる

```python
base = 10
height = '5'

try:
    result = calc_tri(base, height)
    print('計算結果:{}'.format(result))
except Exception as e:
    print('計算できません')
    print(e)
    
```

特定のエラーごとに処理を実行することもできる
finally:→必ず処理が実行されるため、必ず実行したい処理がある時に記載する

```python
try:
    a = 10
    b = 0
    result = a / b
except ZeroDivisionError:
    print('0で割っています')
finally:
    print('ここは必ず通ります')
```
