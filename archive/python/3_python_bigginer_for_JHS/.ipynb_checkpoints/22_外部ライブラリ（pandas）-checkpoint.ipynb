ポイント
DataFrameとSeriesではそれぞれ違うメソッドを持っていることからどちらも覚えておく必要がある

```python
df = pd.read_csv('sample_data/california_housing_test.csv')
```

df.head()：先頭の指定した行数分を表示させる
デフォルトでは5行が出力される

```python
df.head()
```

df.tail()：最終行から指定した行数分を表示させる。
デフォルトでは5行

```python
df.tail()
```

カラム名を指定することで、そのカラムのデータのみを取り出すことができる

```python
df[['longitude', 'latitude', 'housing_median_age']].head()
```

df.iloc[行,列]：指定した行のみを取り出すことができる→アイルック関数と読む
指定の仕方
0:3→0列目から3列目の一つ手前までを取り出すことを意味する

```python
df.iloc[:,0:3].head()
```

一番最後以外を取り出す場合

```python
df.iloc[:,:-1].head()
```

条件付きの抽出の方法

```python
df[df['housing_median_age'] > 30].head()
```

ソートの掛け方
sort_values(by='カラム名')でカラム名で指定したカラムの値でソートをかけることができる
, ascending=False：降順に変更される

```python
df.sort_values(by='total_bedrooms').head()
```

行数と列数を確認できる

```python
df.shape
```

各列ごとの平均値を算出

```python
df.mean()
```

各列ごとの標準偏差を算出

```python
df.std()
```

各列毎の標準偏差や平均値、最大値、最小値などさまざまな値を一度に確認できる

```python
df.describe()
```

NumPyの型に変換することができる

```python
df.values
```
