```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('sample_data/california_housing_test.csv')
```

```python
df.head()
```

scatter：散布図を確認することができる
データの関係性を確認する時に用いる
scatter(横軸, 縦軸)

```python
plt.scatter(df['latitude'],df['longitude'])
```

hist：ヒストグラムを作成することができる
データの散らばりを確認することができる
bins：値を大きくするとより細かくヒストグラムが分かれていく、デフォルトは10くらい

```python
plt.hist(df['latitude'],bins=20)
```

```python
import numpy as np
```

linspace(A, B, C)：A（数字）からB（数字）までをC（数字）等分して配列で出力する

```python
x = np.linspace(1,100,100)
x
```

```python
y = 2*x
y
```

plt.plot(A, B):縦軸B、横軸A折れ線グラフを作成することができる

```python
plt.plot(x, y)
```

```python
y = 1/x
y
```

```python
plt.plot(x, y)
```

```python
y = 5*x*x
y
```
