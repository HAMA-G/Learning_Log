引数→外から持ってくる値のことで下記のものだと()内のnameがそれにあたる

```python
def say_hello2(name):
    print('{}さん、こんにちは。'.format(name))
```

```python
say_hello2('今西')
```

```python
say_hello2('佐藤')
```

```python
def calc_tri(base, height):
    return base * height / 2
```

```python
calc_tri(10,20)
```

デフォルトの値を設定することが可能
エラーを防ぐことができる

```python
def calc_tri(base=10, height=10):
    return base * height * 0.5
```

```python
calc_tri()
```
