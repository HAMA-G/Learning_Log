```python
def say_something():
    message='こんにちは、私は今西です'
    print(message)
```

```python
say_something()
```

messageは関数の中でしか動作しないローカル変数になるためエラーとなる

```python
print(message)
```

```python
message1 = 'こんばんは。私は今西です。'
```

```python
def say_something1():
    print(message1)
```
