__init__　→初期設定を意味する
クラスの中で定義されている変数を「属性」と呼ぶ
クラスの中で定義されている関数のことを「メソッド」と呼ぶ
selfは呼び出したインスタンス自身を指す

```python
class Person:
    #初期設定
    def __init__(self, name, nationality, age):
        self.name = name
        self.nationality = nationality
        self.age = age

    def __call__(self):
        print('ここはcall関数です。')

    def say_hello(self, name):
        print('こんにちは、{}さん。私は{}です。'.format(name, self.name))
```

```python
#インスタンス化（実体化)
imanishi = Person('今西', '日本', '25')
```

name属性

```python
imanishi.name
```

```python
imanishi.say_hello('佐藤')
```

下記のcall関数の呼び出し方は正しくない
特殊な呼び出し方がある

```python
imanishi.__call__()
```

call関数の呼び出し方法

```python
imanishi()
```

```python
imanishi.nationality
```
