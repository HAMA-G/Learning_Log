Syntax Error

構文エラーの際に起こるエラー
つまりPythonの文法的に間違っているときに起こる

NameErro

定義していない変数名を使用した際に起こるエラー
定義し忘れやスペルミスで起こる

AttributeError`

オブジェクトが持っていない属性を使用した際に起こるエラー
例えばタプルが持っていないattend()関数を使用するなど

TypeError

誤ったデータ同士の計算を行ってしまう

ValueError

誤った値を取ってくる場合に起こるエラー
データ型はあっているが、値が間違っているときに起きるエラー

INdentationError

インデント周りで起こるエラー
インデントが上とあっていなかったり、インデントがなかったりする際に起こるエラー

IndexError

範囲外のインデックス（要素番号）を指定した際に起こるエラー
リストの長さが3つなのに4番目の値をとりだそうとしたりすると起こる

KeyError

辞書で登録されていないキーをしていした　際に起こるエラー

ModuleNotFoundError

importで指定したモジュールがない場合に起こるエラー。
インストールしてからimportを行えばOK

FileNotFoundError

指定したファイルが存在しない際に起こるエラー
ファイルが存在しないかファイル名のスペルミス、指定フォルダを間違えている場合に起きる
