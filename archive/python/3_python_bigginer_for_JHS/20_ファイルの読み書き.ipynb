```python
path = 'sample_data/README.txt'
```

w:writeの略で書き込みを意味する

```python
f = open(path, 'w')
```

f.write():ファイルの書き込み

```python
f.write('こんにちは。今西です。')
```

flush：書き込んだ内容を反映させる

```python
f.flush()
```

close:閉じる

```python
f.close()
```

より簡略化する方法

with:ファイルを一時開くことでき、処理が終わると自動で閉じること

```python
with open(path, 'w') as f:
    f.write('おはようございます。今西です。')
```
