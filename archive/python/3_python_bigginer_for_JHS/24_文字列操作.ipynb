埋込

```python
'私の名前は'+name1+'、下の名前は'+name2+'です'
```

上記でも対応できるが、結構毎回記載しないといけないのが面倒

```python
'私の名前は{}、下の名前は{}です'.format(name1, name2)
```

```python
'私の名前は{}、下の名前は{}です。私は{}です。'.format(name1, name2, '25歳')
```
