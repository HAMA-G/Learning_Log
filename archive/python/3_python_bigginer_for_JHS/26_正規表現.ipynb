re:regexの略で正規表現に関するライブラリのこと

```python
import re
```

```python
sentence = 'aaa abc aac abb ababa 123 112 111'
```

```python
pattern = 'ab.'
```

martch:文字列の先頭がマッチするかチェック、抽出
search():先頭に限らずマッチするかチェック、抽出
findall():マッチする部分全てをリストで取得

ab.の正規表現の意味：ab○を抽出する→○は任意の一文字を表現している

```python
results = re.findall(pattern, sentence)

results
```
