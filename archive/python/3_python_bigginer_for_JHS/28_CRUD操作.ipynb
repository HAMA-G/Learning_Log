```python
import sqlite3
import pandas as pd
```

```python
df = pd.read_csv('sample_data/california_housing_test.csv')
```

```python
df.head()
```

```python
db_name = 'sample.db'
conn = sqlite3.connect(db_name)
```

```python
conn
```

```python
df.to_sql('wine_class', conn, if_exists='replace') 
```

```python
c = conn.cursor()
```

```python
query_create = 'select count(*) from wine_class'
```

```python
c.execute(query_create)
```

```python
c.fetchone()
```

insert into wine_class (〇〇, △△）:どのカラムに値を追加するのかを指定する
values (〇〇, △△):追加したい値を格納する

```python
query_create = '''
insert into wine_class (housing_median_age, total_rooms)
values (38.0, 101.0)

'''
```

実行

```python
c.execute(query_create)
```

```python
c.execute(query)
```

```python
c.fetchone()
```

LIMIT：取得するデータ数を指定
OFFSET：取得する行数を指定

```python
query_select = '''
select * from wine_class
LIMIT 5
OFFSET 3000
'''
```

```python
for row in c.execute(query_select):
    print(row)
```

R:Read

```python
query = 'select housing_median_age from wine_class'
```

```python
c.execute(query)
```

```python
print(c.fetchall())
```

```python
query = '''
select housing_median_age from wine_class
where housing_median_age >= 50
'''
```

```python
c.execute(query)
```

```python
c.fetchall()
```

U:Update

```python
query = '''
update wine_class set housing_median_age = 2
where total_rooms >=1000
'''
```

```python
c.execute(query)
```

```python
query = '''
select housing_median_age from wine_class
where housing_median_age  >= 40.0
'''
```

```python
c.execute(query)
```

```python
c.fetchall()
```

D:Delet

```python
query = 'select count(*) from wine_class'
c.execute(query)
c.fetchone()
```

```python
query_delete = '''
delete from wine_class
where total_rooms = 101.0
'''
```

```python
c.execute(query_delete)
```

```python
c.execute(query)
```
