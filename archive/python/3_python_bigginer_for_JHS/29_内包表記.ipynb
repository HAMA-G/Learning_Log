内包表記：リストから何かしらの値を追加していく際、一行で実装できる構文構造

```python
import random
```

```python
random.randint(1, 100)
```

```python
data = []
for _ in range(1000):
    x = random.randint(1, 100)
    data.append(x)
```

```python
data
```

```python
len(data)
```

```python
%%timeit
data_10x = []

for datum in data:
    data_10x.append(datum*10)
```

```python
data_10x[:10]
```

上記を内包表記で記載する
[]の中にfor文を記載する
forの前にリストに追加したいデータの内容を記載する
ワンライン記法という
利点：1行で記載できてみやすい、処理速度が速い
%%timeit：処理速度を測定できるマジックコマンド

```python
%%timeit
new_data_10x = [datum*10 for datum in data]
```

```python
new_data_10x[:10]
```

```python
data_odd = []
for datum in data:
    if datum % 2 != 0:
        data_odd.append(datum)
```

上記のif文を用いた内包表記の記載方法

```python
new_data_odd = [datum for datum in data if datum % 2 != 0]
```

```python
new_data_odd[:10]
```

50以上の整数→そのまま表示
50未満の整数→100倍

```python
data_50 = []
for datum in data:
    if datum >= 50:
        data_50.append(datum)
    else:
        data_50.append(datum*100)
```

```python
data_50[:10]
```

if文の中でelseを用いた際の内包表記

```python
new_data_50 = [datum if datum >= 50 else datum*100 for datum in data]
```

FizzBuzzm問題
1から50までの数字を出力するプログラミングを書いてください。
ただし、数字の3の倍数の時は数字の代わりにFizzと出力し
5の倍数の時は数字の代わりにBuzzと出力し
3と5の倍数の時は、FizzBuzzと出力すること。

```python
fizz_buzz = []
for i in range(1, 51):
    if i % 15 == 0:
        fizz_buzz.append('FizzBuzz')
    elif i % 3 == 0:
        fizz_buzz.append('Fizz')
    elif i % 5 == 0:
        fizz_buzz.append('Buzz')
    else:
        fizz_buzz.append(i)
```

```python
fizz_buzz
```

上記の内包表記
基本的には上から順番に書いていけばいい

```python
new_fizz_buzz = ['FizzBuzz' if i % 15 == 0 else 'Fizz' if i % 3 == 0 else 'Buzz' if i % 5 == 0 else i for i in range(1, 51)]
```

```python
new_fizz_buzz
```
