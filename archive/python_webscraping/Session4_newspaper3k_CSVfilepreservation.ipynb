```python
import newspaper
import csv
import datetime
```

```python
url = "https://www.bloomberg.co.jp/"
```

build：サイトのトップページからh区数のページをまとめて取得することが可能
memoize_articles=False：コードの再実行ごとに、記事データを再取得する
article.download：articleに対してdownloadで記事のHTMLデータをダウンロードする
article.parese：取得したHTMLデータの解析し、記事のタイトルや本文アドのデータを拾ってくる
article.title：取得した記事のタイトル
article.url：取得した記事のURL
article.text：記事の文章
end="¥n¥n"
今回は10件の記事を取得するためif文を作成

```python
website = newspaper.build(url, memoize_articles=False)

article_list = []

i = 1
for article in website.articles:
    article.download()
    article.parse()

    # print(f"記事 {i}:", article.title)
    # print(f"記事 {i}:{article.title}")
    # print(article.url)
    # print(article.text, end="¥n¥n")

    article_list.append([str(i), article.title, article.url, article.text])

    if i > 9:
        break
    # i = i + 1
    i += 1
```

Pythonのf文字列
文字列の前にfをつけることで、文字列のなかに含まれる置換フィールド{}に変数を指定して出力することが可能
例）
test = 999
print(f'変数testの値は、{test}です。')
→実行した際の出力は「変数testの値は、999です。」になる。

```python
csv_header = ["記事番号", "タイトル", "URL", "本文"]
csv_date = datetime.datetime.today().strftime("%Y%m%d%H")
csv_file_name = "bloomberg_" + csv_date + ".csv"
with open(csv_file_name, "w", errors = "ignore") as f:
    writer = csv.writer(f, lineterminator = "\n")
    writer.writerow(csv_header)
    writer.writerows(article_list)
```

strftime→日付を文字列に変換する
日付.strftime(文字列の書式)
文字列の書式の記載は必須下記のように日付を指定できる
