```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.jashin.info/kiura/scraping_basic/01basic_tag.html"
```

```python
res = requests.get(url)
```

```python
res.encoding = "utf-8"
```

```python
soup = BeautifulSoup(res.text, "html.parser")
```

```python
titles = soup.find_all(class_="author")
```

```python
for title in titles:
    print(title.text)
```

```python
titles = soup.find(class_="author")
```

```python
for title in titles:
    print(title.text)
```
