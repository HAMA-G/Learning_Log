```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.jashin.info/kiura/scraping_basic/02links.html"
```

```python
res = requests.get(url)
```

```python
res.encoding = "utf-8"
```

```python
soup = BeautifulSoup(res.text, "html.parser")
```

```python
links = soup.find_all("a")
```

```python
for link in links:
    try:
        print(link.text)
        print(link["href"])
    except:
        print("リンク切れ")
```

```python
titles = soup.find(class_="author")
```

```python
for title in titles:
    print(title.text)
```
