```python
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.jashin.info/kiura/scraping_basic/03css_selectors.html"
```

```python
res = requests.get(url)
```

```python
res.encoding = "utf-8"
```

```python
soup = BeautifulSoup(res.text, "html.parser")
```

```python
science = soup.select('.genre[data-name="science"] .author')
```

```python
for sciences in science:
    print(sciences.text)
```
